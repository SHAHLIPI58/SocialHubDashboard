```python

chicago_downtown_zipcodes = [60611,60616,60612,60601, 60606,60607, 60608,60610,60602,60603, 60604, 60605]

for zip_code in chicago_downtown_zipcodes:
    for x in range(20):
        response = yelp_api.search_query(categories='bowling+bars+restaurants+parks+movietheaters', 
                                     location=zip_code, 
                                     
                                 sort_by='rating', limit=50, offset=x*50)
        df__business_reviews = df__business_reviews.append(pd.DataFrame(response['businesses']))
```

```python
lat = []
lon = []

for value in df__business_reviews['coordinates']:
        lat.append(value['latitude'])
        lon.append(value['longitude'])



df__business_reviews['latitude']=lat
df__business_reviews['longitude']=lon
```

```python
cate = []
for value in df__business_reviews['categories']:
    temp=[]
    for cat in value:
        temp.append(cat['alias'])
        temp.append(cat['title'])
    cate.append(temp)

df__business_reviews['CategoriesList']=cate
```

```python
address = []

for value in df__business_reviews['location']:
    if len(value['display_address']) > 0:
        address.append(' '.join(map(str, value['display_address'])))
    else:
        address.append("")
        
df__business_reviews['address']=address
```
